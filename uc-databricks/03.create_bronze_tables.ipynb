### Create Bronze Tables
1. clientes.csv from 'abfss://bronze@sadatabricksexternal.dfs.core.windows.net/clientes.csv'
2. detalles.csv from 'abfss://bronze@sadatabricksexternal.dfs.core.windows.net/detalles.csv'

```sql
DROP TABLE IF EXISTS uc_bricks.bronze.clientes; --The DROP TABLE statement is useful for subsequently adding all the notebooks to the same workflow

CREATE TABLE IF NOT EXISTS uc_bricks.bronze.clientes(
  clienteId INT,
  nombre STRING,
  paterno STRING,
  materno STRING,
  documentoIdentidad STRING,
  telefono STRING,
  email STRING,
  direccion STRING
)
USING CSV
OPTIONS (path 'abfss://bronze@sadatabricksexternal.dfs.core.windows.net/clientes.csv', delimiter ';', header 'true')
```

```sql
DROP TABLE IF EXISTS uc_bricks.bronze.detalles;

CREATE TABLE IF NOT EXISTS uc_bricks.bronze.detalles(
  detalleId INT,
  clienteId INT,
  servicio STRING,
  plan STRING,
  precioMensual DECIMAL(5,2),
  estado BOOLEAN,
  fechaInicio DATE,
  fechaFin DATE
)
USING CSV
OPTIONS (path 'abfss://bronze@sadatabricksexternal.dfs.core.windows.net/detalles.csv', delimiter ';', header 'true')
```
