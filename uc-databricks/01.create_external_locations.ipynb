### Create External Locations
1. Bronze
2. Silver
3. Gold

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS external_bronze
URL 'abfss://bronze@sadatabricksexternal.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `databricks-ext-storage-credential`);
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS external_silver
URL 'abfss://silver@sadatabricksexternal.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `databricks-ext-storage-credential`);
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS external_gold
URL 'abfss://gold@sadatabricksexternal.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `databricks-ext-storage-credential`);
```
