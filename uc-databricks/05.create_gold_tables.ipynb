### Create Gold Managed Tables
- Using JOIN on 'clientes' and 'detalles'

```sql
DROP TABLE IF EXISTS uc_bricks.gold.ingreso_cliente;

CREATE TABLE IF NOT EXISTS uc_bricks.gold.ingreso_cliente
AS
  SELECT C.nombre_completo, SUM(D.precio_mensual) AS ingreso
  FROM uc_bricks.silver.clientes AS C
  INNER JOIN uc_bricks.silver.detalles AS D
    ON C.cliente_id = D.cliente_id
  WHERE D.estado = true
  GROUP BY C.nombre_completo
  ORDER BY ingreso DESC;
```
