### Create Silver Managed Tables
1. clientes
2. detalles

```sql
DROP TABLE IF EXISTS uc_bricks.silver.clientes;

CREATE TABLE IF NOT EXISTS uc_bricks.silver.clientes
AS
  SELECT
  clienteId AS cliente_id,
  CONCAT(nombre, ' ', paterno, ' ', materno) AS nombre_completo,
  documentoIdentidad AS documento_identidad,
  telefono,
  email,
  direccion,
  current_timestamp() AS ingestion_date
  FROM uc_bricks.bronze.clientes;
```

```sql
DROP TABLE IF EXISTS uc_bricks.silver.detalles;

CREATE TABLE IF NOT EXISTS uc_bricks.silver.detalles
AS
  SELECT
  detalleId AS detalle_id,
  clienteId AS cliente_id,
  servicio,
  plan,
  precioMensual AS precio_mensual,
  estado,
  fechaInicio AS fecha_inicio,
  fechaFin AS fecha_fin,
  current_timestamp() AS ingestion_date
  FROM uc_bricks.bronze.detalles;
```
